```python
from langchain.llms import HuggingFacePipeline
from transformers import LlamaTokenizer, LlamaForCausalLM, pipeline  , GenerationConfig , BitsAndBytesConfig
import torch
from langchain import PromptTemplate, LLMChain
```

```python
MODEL_ID = "chavinlo/alpaca-native"
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True
)
tokenizer = LlamaTokenizer.from_pretrained(MODEL_ID)
model = LlamaForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config=quantization_config,
    device_map="auto"
)
```

```python
hf_pipeline=pipeline('text-generation',model=model,tokenizer=tokenizer,max_new_tokens=3000)
llm=HuggingFacePipeline(pipeline=hf_pipeline)
```
